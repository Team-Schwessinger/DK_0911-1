### Allelic variation analysis in DK0911 and Pst104E

This notebook started as a draft analysis of the allele/gene pair variation between psuedo haplotypes and in one isolate vs the same observations in a different isolate.  

Mostly this whole notebook is still here for record purposes. At the end I only used high quality allele pairs for the analysis. Here high quality allele pairs are pairs that are located on orthologous contigs that are linked within FALCON-Unzip. Hence for this analysis it doesn't matter if the haplotype blocks are correctly phased or not.

```python
def ranksumtest(df1, df2, matchType, to_test):
    return stats.ranksums\
    (df1.loc[df1.matchType == matchType, to_test],df2.loc[df2.matchType == matchType, to_test] )
```
